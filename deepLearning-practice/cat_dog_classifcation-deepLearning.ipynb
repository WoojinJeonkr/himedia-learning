```python
# CNN 특성추출을 하려면 이미지가 4차원이 되어야 한다!
# (4000, 150, 150, 3)
# 4000: 이미지 개수
# 150, 150: 가로세로 픽셀
# 3: 칼라
# 0으로 채워지는 4차원 배열을 하나 만들어두었다가
# 이미지 하나하나를 스케일링하고 리사이징해서
# 4차원에 하나씩 넣음
```

```python
total_images = np.zeros((5, data_height, data_width, channel_n))
```

```python
# 읽어온 이미지 목록을 넣으면 라벨링/이미지 스케일링/리사이징해서
# 우리가 원하는 이미지 데이터 생성
def make_file(data_height, data_width, channel_n, data_list_n):
    
    # 라벨링한 결과 넣어줄 1차원 리스트
    label = []
    
    # 4차원으로 만들어줄 4차원 리스트
    images = np.zeros((data_list_n, data_height, data_width, channel_n))
    
    # 이미지 리스트에서 하나씩 꺼내서 라벨링, 스케일링, 리사이징
    for i, image in enumerate(data_list):
        
        # 라벨링: dog.01.jpg
        token = text_to_word_sequence(image)
        label.append(token[0])
        
        # 데이터를 읽어서, 스케일링, 리사이징
        image2 = cv2.imread(image)
        image3 = cv2.resize(image2, (data_height, data_width)) / 255
        
        # 4차원 리스트에 하나씩 넣음
        images[i, :, :, :] = image3
    label2 = np.array(label)
    return (label2, images)
```

```python
labels, images = make_file(data_height, data_width, channel_n, 4000)
```

```python
labels
```

```python
labels.shape
```

```python
items = labels
encoder = LabelEncoder()
label = encoder.fit_transform(items)
label, len(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = images
y = label
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    test_size= 0.2,
                                                    random_state=55
                                                   )
print(X_train.shape, X_test.shape)
```

```python
## 딥러닝 모델 생성
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
# 특성 추출 레이어 쌓기
model.add(Conv2D(16, (3, 3), activation='relu', input_shape=(data_height, data_width, 3)))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(2048, (3, 3), activation='relu'))
model.add(MaxPool2D((2,2)))

# 딥러닝 학습 레이어 쌓기
# 입력 -- 히든 -- 출력 순서대로 쌓기

# 입력 -- 1차원이어야 함
model.add(Flatten())

# 히든
model.add(Dense(1024, activation='relu'))
model.add(Dropout(rate=0.8))
model.add(Dense(512, activation='relu'))
model.add(Dropout(rate=0.8))
model.add(Dense(128, activation='relu'))
model.add(Dropout(rate=0.8))
model.add(Dense(128, activation='relu'))
model.add(Dropout(rate=0.8))
model.add(Dense(32, activation='swish'))
model.add(Dropout(rate=0.5))
model.add(Dense(16, activation='swish'))
model.add(Dropout(rate=0.5))
model.add(Dense(8, activation='swish'))
model.add(Dropout(rate=0.5))

# 출력(1개) -- 딥러닝 학습의 종류에 따라 다르게 써주어야 함!!
## 이진 분류인 경우 (1, sigmoid)
## 다중 분류인 경우 (n, softmax)
## 회귀인 경우(1, linear)
model.add(Dense(1, activation='sigmoid')) #결과가 엘리먼트로 출력됨. 0.5를 기준으로 판별
model.summary()
```

```python
# 모델의 설정값 지정
# 1. 기울기 찾는데 최적의 방법 설정(optimizer)
# 2. 특정 가중치를 가지는 경우 오차률 계산 방법 설정(loss)
#    회귀인 경우: mae
#    분류인 경우: 이진(binary_crossentropy),
#                 다중(sparse_categorical_crossentropy)
# 3. fit시킬 때 fit시키는 상황이 순서대로 프린트되는데 어떤 항목이 출력되는 항목 설정(metrics)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy']
)
```

```python
# 모델 학습
## epoch마다 train의 loss, acc뿐만 아니라
##           test의 loss, acc를 알고 싶다면
##           fit(validation_data = (X_test, y_test))
## 조기 종료(earlyStopping): epoch를 실행할 때 최적의 acc인 경우, 더이상 epoch를 수행하지 않아도 되게 설정
## 모델 저장(ModelCheckPoint): 최적의 acc인 경우의 가중치를 자동으로 모델 파일로 저장하게 설정
## 특정한 상황을 기다렸다가 특정한 처리를 하고자 하는 경우(callback함수)
## 1) 객체를 만들어주고, 2) fit(callbacks = [객체]) 
es = EarlyStopping(patience=3, restore_best_weights=True)
cp = ModelCheckpoint('best_cnn_1.h5')
cnn_history = model.fit( X_train, 
                         y_train,                
                         batch_size=32,
                         epochs=20,
                         validation_data = (X_test, y_test),
                         callbacks = [es, cp],
                         verbose=2
                        )
```

```python
test_result = model.evaluate(X_test, y_test)
test_result
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1) ## 1행2열의 도표, 1열
plt.plot(cnn_history.history['accuracy'])
plt.plot(cnn_history.history['val_accuracy'])
plt.title('model acc')
plt.xlabel('epoch')
plt.ylabel('acc')
plt.legend(['train_acc','test_acc'])

plt.subplot(1, 2, 2) ## 1행2열의 도표, 2열
plt.plot(cnn_history.history['loss'])
plt.plot(cnn_history.history['val_loss'])
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train_loss','test_loss'])
```
