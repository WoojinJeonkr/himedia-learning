### 012. 사용자 지정 라벨

```python
# 지정하고자 하는 라벨의 형태를 함수로 정의
def summary(values, ranks):
    total_deaths = int(round(values.sum(), -2))
    s = f'Total Deaths - {total_deaths:,.0f}'
    return {'x': .99, 'y': .05, 's': s, 'ha': 'right', 'size': 8}

bcr.bar_chart_race(covid, period_summary_func = summary)
```

```python
# 동영상으로 저장
bcr.bar_chart_race(covid, period_summary_func = summary, filename = "covid_period_summary.mp4")
```

### 014. 서브 플롯으로 표시

```python
import matplotlib.pyplot as plt
from matplotlib import dates

fig, ax_array = plt.subplots(2, 2, figsize=(8, 4), dpi=120, tight_layout=True)
ax1, ax2, ax3, ax4 = ax_array.flatten()
fig.suptitle('Animation in First Axes', y = 1) # 모든 서브 플롯에 하나의 제목으로 정의

ax2.plot(covid) # 2번째 플롯에 선형 그래프로 표시
ax2.xaxis.set_major_locator(dates.DayLocator([3, 7, 12]))
ax3.bar(covid.index, covid.median(axis = 1)) # 3번째 플롯에 막대 그래프로 표시(중앙값)
ax3.xaxis.set_major_locator(dates.DayLocator([3, 7, 12]))
ax4.pie(covid.iloc[-1], radius = 1.5, labels = covid.columns) # 4번째 플롯에 원 그래프로 표시

bcr.bar_chart_race(covid, n_bars = 3, fig = fig) # 비어있는 플롯인 1번째 플롯에 움직이는 막대 그래프로 표시
```

## 05. 움직이는 막대 그래프 동영상으로 저장하기

```python
bcr.bar_chart_race(
    df = covid, # 데이터
    filename = 'covid19_horiz.mp4', # 파일 이름
    orientation = 'h', # 그래프의 방향(h: 수평, v: 수직)
    sort = 'desc', # 막대 정렬 순서
    n_bars = 6, # 표시할 막대 개수
    fixed_order = False, # 위에서부터 보여질 라벨 순서
    fixed_max = True, # x축 또는 y축 최대값을 데이터의 최대값으로 고정
    steps_per_period = 10, # 화면전환 프레임수 결정(값이 높을수록 부드럽게 움직인다)
    interpolate_period = False, # 한 프레임당 증가하는 시간을 선형적으로 조절
    label_bars = True, # 막대 라벨 표시 여부
    bar_size = .95, # 막대 크기
    period_label = {'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'}, # 기간 라벨 속성 
    period_fmt = '%B %d, %Y', # 기간 라벨 표시 형식
    period_summary_func = lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total deaths: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'}, # 데이터값 표시 형식
    perpendicular_bar_func = 'median', # 데이터의 통계값의 중간값을 수직선으로 표현
    period_length = 500, # 주기 조절(값이 높을 수록 느리다)
    figsize = (5, 3), # 그래프가 그려지는 크기
    dpi = 144, # dpi 설정
    cmap = 'dark12', # 색상 지정
    title = 'COVID-19 Deaths by Country', # 제목
    title_size = '', # 제목 크기
    bar_label_size = 7, # 막대 라벨 크기
    tick_label_size = 7, # 눈금 라벨 크기
    shared_fontdict = {'family' : 'Helvetica', 'color' : '.1'}, # 플롯의 모든 텍스트에 대한 공통 형식 지정
    scale = 'linear', # 눈금 단위 지정
    writer = None, # 애니메이션 writer 설정
    fig = None, # matplotlib figure
    bar_kwargs = {'alpha': .7}, # 막대 설정
    filter_column_colors = False # 화면에 표시되는 막대에만 색상을 적용할 지에 대한 여부(적용 시에는 True로 작성)
)
```
