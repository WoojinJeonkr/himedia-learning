```python
# 1. 크롬 브라우저를 실행하자
def connect():
    driver = wd.Chrome('/Users/hi/Desktop/python_project/chromedriver.exe')
    return driver
```

```python
driver2 = None
url = 'https://www.coffeebeankorea.com/store/store.asp'
```

```python
# 2. 특정 페이지를 가지고 와서, 스크립트를 실행하자
def coffee_bean(no):
    driver2.get(url)
    time.sleep(1)
    func = "storePop2('" + str(no) + "')"
    shop = []
    try:
        # 스크립트 실행
        driver2.execute_script(func)
        time.sleep(1)
        
        # 실행한 결과 html 가지고 오세요.
        html2 = driver2.page_source
        
        # bsparser설정
        doc2 = BeautifulSoup(html2, 'html.parser')
        
        # 타이틀 검색
        title3 = doc2.select('div.store_txt > h2')
        print('title3>> ', title3)
        title4 = title3[0].string
        print('title4>> ', title4)
        shop.append(title4)
        
        # 주소 검색
        info2 = doc2.select('div.store_txt > table.store_table > tbody > tr > td')
        addr3 = info2[2]
        addr4= addr3.getText()
        shop.append(addr4)

        # 전화번호 검색
        tel3 = info2[3]
        tel4 = tel3.string
        shop.append(tel4)
        
    except:
        print('해당번호없음')

    return shop #[삼성봉은사거리점,서울시 강남구 영동대로 607 1,2층,02-3443-5618]
```

```python
driver2 = connect()
```

```python
coffee_bean(31) # return shop list
```

```python
for i in range(1, 6):
    print('-------',i,'--------')
    shop3 = coffee_bean(i)
    print(shop3)
```

```python
title_list2= []
addr_list2 = []
tel_list2 = []
```

```python
for i in range(1, 300):
    print('-------',i,'--------')
    shop3 = coffee_bean(i)
    if len(shop3) != 0:
        title_list2.append(shop3[0])
        addr_list2.append(shop3[1])
        tel_list2.append(shop3[2])
    print(shop3)
```
