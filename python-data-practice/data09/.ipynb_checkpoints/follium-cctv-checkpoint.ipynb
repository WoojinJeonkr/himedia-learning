```python
import pandas as pd
```

```python
df = pd.read_csv('../csv-data/folium-data.csv', encoding='utf-8')
df.head()
```

```python
del df['no']
```

```python
df.head()
```

```python
df.set_index('자치구', inplace=True)
df.head()
```

```python
# 대기업
map = folium.Map(location=[37.57, 126.98],
                 zoom_start=11
                )
map.choropleth(geo_data = geo_str,
               data = df['대기업'],
               columns=[df.index, df['대기업']],
               fill_color='PuRd',
               key_on = 'feature.id'
        )
map
```

```python
# 중기업
map = folium.Map(location=[37.57, 126.98],
                 zoom_start=11
                )
map.choropleth(geo_data = geo_str,
               data = df['중기업'],
               columns=[df.index, df['중기업']],
               fill_color='PuRd',
               key_on = 'feature.id'
        )
map
```

```python
# 소기업
map = folium.Map(location=[37.57, 126.98],
                 zoom_start=11
                )
map.choropleth(geo_data = geo_str,
               data = df['소기업'],
               columns=[df.index, df['소기업']],
               fill_color='PuRd',
               key_on = 'feature.id'
        )
map
```

```python
# 서클 그리기
map = folium.Map(location=[37.57, 126.98],
                 zoom_start=11)

map.choropleth(geo_data = geo_str,
               data = df['대기업'],
               columns = [df.index, df['대기업']],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on = 'feature.id')

for n in df.index:
    folium.CircleMarker([df['lat'][n], df['lng'][n]], 
                        radius = df['대기업'][n]*0.1, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map)
map
```

```python
map.save('follium-company.html') # 파일의 모든 map 이미지가 아닌 선택한 map 이미지만 저장된다
```

```python
# cctv파일 열어서 2개 이상의 항목으로 folium으로 그리기
```

```python
cctv = pd.read_csv('../data06/cctv_data_collect/crime_seoul_3.csv', encoding='euc-kr')
cctv.head()
```

```python
cctv.rename(columns={'구별':'자치구'}, inplace=True)
cctv.head()
```

```python
geo_df = pd.merge(df, cctv, on='자치구')
```

```python
geo_df.head()
```

```python
del geo_df['소상공인']
del geo_df['소기업']
del geo_df['중기업']
del geo_df['대기업']
```

```python
geo_df.head()
```

```python
geo_df.set_index('자치구', inplace=True)
geo_df.head()
```

```python
geo_df.reset_index(inplace=True)
```

```python
geo_df.drop([1, 3, 10, 15, 26, 28], inplace=True)
geo_df.info()
```

```python
geo_df.duplicated().sum()
```

```python
geo_df.set_index('자치구', inplace=True)
```

```python
# 종로구의 살인 발생
map = folium.Map(location=[37.57, 126.98], 
                zoom_start=10
                )
map.choropleth(geo_data=geo_str,
                      data= geo_df['살인 발생'],
                      columns=[geo_df.index, geo_df['살인 발생']],
                      fill_color='PuRd',
                      key_on='feature.id'
              )


for n in geo_df.index:
    folium.CircleMarker([geo_df['lat'][n], geo_df['lng'][n]], 
                        radius = geo_df['살인 발생'][n]*0.1, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map)
map
```
