```python
## Decision Tree 비교하기
from sklearn.tree import DecisionTreeClassifier
```

```python
# max_depth가 5, 10, 15일 경우를 비교해주고 싶은 경우
tree_clf = DecisionTreeClassifier(
                max_depth=15, random_state=42
)
```

```python
train_exam, test_exam, train_target, test_target = train_test_split(
                mnist['data'],  # 문제(DataFrame)
                target, # 정답(ndarray, Series, list와 같은 1차원 데이터)
                test_size=0.2, # 검증비율 20%, 훈련비율 80%
                # stratify=True, # 타겟의 비율을 골고루 맞춰주는 역할
                random_state=100 # 시드값
                )
```

```python
train_exam.shape
```

```python
test_exam.shape
```

```python
train_target.shape
```

```python
test_target.shape
```

```python
type(test_target)
```

```python
tree_clf.fit(train_exam, train_target)
```

```python
tree_clf.score(train_exam, train_target)

# max_depth = 3 의 경우
# 0.45803571428571427 --> 45.8% --> 검증 불가
# 최소 55% 이상은 되어야 검증이 가능하다

# max_depth = 10 의 경우
# 0.8996428571428572 --> 89.9%

# max_depth = 15 의 경우
# 0.9819821428571428 --> 98.2% 
```

```python
####### max_depth 찾아보기 ########
```

```python
# max_depth가 5, 10, 15일 경우를 비교해주고 싶은 경우
train_acc = []
test_acc = []
for md in [5, 10, 15]:
    tree_clf = DecisionTreeClassifier(
                    max_depth=md, random_state=42
    )
    tree_clf.fit(train_exam, train_target) # 훈련데이터를 가지고 공부
    score = tree_clf.score(train_exam, train_target) # 훈련데이터 점수
    train_acc.append(score) # 훈련데이터 점수 누적
    
    result2 = tree_clf.predict(test_exam) # 검증용 데이터 테스트를 봄
    acc = accuracy_score(test_target, result2) # 검증데이터 점수
    test_acc.append(acc) # 검증데이터 점수 누적
```

```python
# 훈련데이터 점수와 검증 데이터 점수가 비슷하게 나오거나 
# 검증데이터 점수가 높은 모델이 좋은 모델이다
print('train_acc >> ', train_acc)
print('test_acc >> ', test_acc)
```
