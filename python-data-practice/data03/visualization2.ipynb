```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['total_bill'], rug=True)
```

```python
tips.describe()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=True)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=True, ci=None)
# ci: 신뢰구간, bootstrapping을 사용하여 계산되며 대규모 데이터셋에 대해 많은 시간이 소요되므로 ci=None을 사용하여 비활성화
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
sns.barplot(x='time', y='tip', data=tips)
```

```python
sns.barplot(x='time', y='tip', data=tips, ci=None)
```
