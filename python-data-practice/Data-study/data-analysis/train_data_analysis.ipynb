```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("../data-analysis/train.csv", encoding='euc-kr')
train
```

```python
train = train.astype({'합 계':'int'}, {'호선':'int'})
```

```python
use_train = train[['호선', '역명', '합 계']]
use_train
```

```python
A = use_train[use_train['호선']==1]
A
```

```python
t_train = []
for i in range(1, 9):
    t_train.append(use_train[use_train['호선']==i]['합 계'].sum())
t_train
```

```python
total_train = pd.DataFrame(data = t_train,
                               index=['1', '2', '3', '4', '5', '6', '7', '8'],
                               columns=['사용한 총 인원'])
total_train
```

```python
total_train.index.name = '호선'
```

```python
total_train.to_csv('호선 별 지하철 사용 총 인원.csv', encoding='utf-8-sig')
```

```python
import seaborn as sns
```

```python
detect = pd.read_csv('./호선 별 지하철 사용 총 인원.csv', encoding='utf-8', engine='python')
```

```python
detect.index.name = '호선'
detect.columns = ['사용한 총 인원']
```

```python
detect
```

```python
ax = detect.plot(kind='bar', title='호선 별 지하철 이용자 현황', figsize=(10,6))
plt.xticks(rotation=0)
for p in ax.patches: # ax가 가리키는 그래프에서 막대들을 담고 있는 리스트만큼 반복
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.1f m"%(height/1e6), (left+width/2, height*1.02), ha='center')
plt.sca(ax)
# plt.box(False) # 박스 보이기, False면 보이지 않음
```

```python
detect['사용한 총 인원'].plot.pie(figsize=(6,8), 
                                  autopct='%1.1f%%') # autopct: 그래프에 값 표시
plt.legend() # 범례 표시
```
