### 전처리

```python
# 제목 추출
all_titles = df1000['제목']
all_titles[:5]
```

```python
stopwords = set(stopwords.words('english'))
```

```python
lemma = WordNetLemmatizer()
```

1. 글자만 추출(글자가 아닌 것은 공백으로 replace)
2. 분석에 사용하지 않을 글자들을 제거 (불용어 처리)
3. 표제어 추출

```python
words = [] # 전처리한 결과를 모으는 리스트
```

```python
# for title in all_titles[0:1]:
#     # 전처리 1, 글자가 아닌 것은 공백으로 대체
#     prel = re.sub(r"[^a-zA-Z]+", " ", str(title))
#     print(prel)
#     # 소문자로 변경하여 단어를 토큰으로 만들자
#     prel_lower_token = word_tokenize(prel.lower())
#     print(prel_lower_token)
#     # 전처리 2. 불용어 처리
#     pre2 = [w for w in prel_lower_token if w not in stopwords]
#     print(pre2)
#     ## 4. 표제어 추출
#     pre3 = [lemma.lemmatize(w) for w in pre2]
#     print(pre3)

# 결과
# Clustering Big Spatiotemporal Interval Data
# ['clustering', 'big', 'spatiotemporal', 'interval', 'data']
# ['clustering', 'big', 'spatiotemporal', 'interval', 'data']
# ['clustering', 'big', 'spatiotemporal', 'interval', 'data']
```

```python
for title in all_titles:
    # 전처리 1, 글자가 아닌 것은 공백으로 대체
    prel = re.sub(r"[^a-zA-Z]+", " ", str(title))
    # 소문자로 변경하여 단어를 토큰으로 만들자
    prel_lower_token = word_tokenize(prel.lower())
    # 전처리 2. 불용어 처리
    pre2 = [w for w in prel_lower_token if w not in stopwords]
    ## 4. 표제어 추출
    pre3 = [lemma.lemmatize(w) for w in pre2]
    words.append(pre3)
```

```python
reduce(lambda x, y : x + y, words[:2]) # 결과가 list
# 여러 개의 list를 하나로 합하고 싶은 경우(요약하고 싶을 때)
```

```python
words2 = reduce(lambda x, y : x + y, words)
print(type(words2))

words2[:10]
```

```python
count = Counter(words2)
count
```

```python
count.most_common(3)
```

```python
for tag, counts in count.most_common(2):
    print(tag, ',', counts)
```

```python
word_count= dict()
for tag, counts in count.most_common(50):
    if(counts > 30): # 빈도수가 30 이상인 것만 추출
        word_count[tag] = counts # 'data' : 1250
print(word_count)
```
