```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
# 나이가 30살 이상 남자 승객 조건 필터링
# fare를 많이 낸 순서로 내림차순 정렬
# 상위 10개를 출력
```

```python
df[df['age'] >= 30] # 결과는 df
```

```python
df[df['age'] >= 30]['age'].value_counts().head()
```

```python
# 나이가 30살 이상 남자 승객 조건 필터링
condition1 = (df['age'] >= 30)
condition2 = (df['who'] == 'man')
df.loc[condition1 & condition2]
```

```python
# fare를 많이 낸 순서로 내림차순 정렬
# 상위 10개를 출력
result = df.sort_values(by='fare', ascending=False).head(10)
result
```

```python
# fare를 많이 낸 순서로 내림차순 정렬된 결과를 인덱스 순서대로 내림차순 정렬
result.sort_index(ascending=False)
```

```python
# 통합
condition1 = (df['age'] >= 30)
condition2 = (df['who'] == 'man')
result1 = df.loc[condition1 & condition2]
result1.sort_values(by='fare', ascending=False).head(10)
```

```python
# 나이가 20살 이상 40살 미만인 승객
# pclass가 1등급 혹은 2등급인 승객
# 열(column)은 survived, pclass, age, fare 만 나오게 출력
# 10개만 출력
```

```python
# 나이가 20살 이상 40살 미만인 승객
condition3 = (df['age'] >=20)
condition4 = (df['age'] < 40)
df.loc[condition3 & condition4]
```

```python
# pclass가 1등급 혹은 2등급인 승객
condition5 = (df['pclass'] == 1)
condition6 = (df['pclass'] == 2)
df.loc[condition5 | condition6]
```

```python
# 열(column)은 survived, pclass, age, fare 만 나오게 출력
# 10개만 출력
df[['survived', 'pclass', 'age', 'fare']][:10]
```

```python
# 통합
conditionA = ((df['age'] >=20)&(df['age'] < 40))
conditionB = (df['pclass'] < 3)
df.loc[conditionA & conditionB, ['survived', 'pclass', 'age', 'fare']][:10]
```
