```python
import pandas as pd
import numpy as np
```

```python
# 1. df로 읽어들이고
df = pd.read_csv('scientists.csv', encoding='utf-8')
df
```

```python
# 2. df의 전체 데이터셋 타입을 확인
df.dtypes
```

```python
df['Born'] = pd.to_datetime(df['Born'])
df['Died'] = pd.to_datetime(df['Died'])
```

```python
df.dtypes
```

```python
# 3. df의 컬럼 리스트를 확인
df.columns
```

```python
# 4. df의 값을 확인
df.values
```

```python
# 5. df의 맨 마지막에 price 항목을 추가(값은 임의대로 넣음, int)
df['price'] = 12667
df
```

```python
# 6. df의 맨 마지막에 percent항목을 추가(값은 임의 대로 넣음, float)
df['percent'] = 0.28
df
```

```python
# 7. df의 맨 마지막에 sale 항목을 추가(price * percent)
df['sale'] = df['price']*df['percent']
df
```

```python
# 8. df의 Name 컬럼 추출
df['Name']
```

```python
# 9. df의 인덱스 확인
df.index
```

```python
# 10. df의 Name 컬럼 값이 Rosaline인 행 추출
df[df['Name'].str.contains('Rosaline')]
```

```python
# 11. df의 Name 컬럼 중 3~5 사이의 행 추출
df.loc[3:5, 'Name']
```

```python
# 12. df의 Name, Died 컬럼 모든 행 추출
df.loc[:,'Name':'Died']
```

```python
# 13. df의 Born 컬럼부터 끝 컬럼까지 모든 행 추출
df.loc[:,'Born':]
```

```python
# 14. df의 Born, Age 컬럼의 5~7행 추출
df.loc[5:7, 'Born':'Age']
```

```python
# 15. df.shape을 통해 몇행 몇열인지 확인
df.shape
```

```python
# 16. df의 숫자 데이터의 상세정보 확인
df.describe()
```

```python
# 17. Nurse라는 스트링 추출(index와 컬럼명 사용)
df.loc[2,'Occupation']
```
