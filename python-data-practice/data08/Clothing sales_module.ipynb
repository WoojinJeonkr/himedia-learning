```python
exam =apparel[['List_price', 'Promotion','Cost','price paid']]
exam.head()
```

```python
target=apparel['subfamily']
target[:2]
```

```python
apparel['subfamily'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_exam, test_exam, train_target, test_target = train_test_split(exam,  
                                                                    target,
                                                                    test_size = 0.2,
                                                                    random_state= 42
                                                                   )
```

```python
train_exam.head()
```

```python
train_exam.reset_index(inplace=True)
del train_exam['index']
train_exam.head()
```

```python
train_target = train_target.reset_index()
del train_target['index']
train_target.head()
```

```python
test_exam.reset_index(inplace=True)
del test_exam['index']
test_exam.head()
```

```python
test_target = test_target.reset_index()
del test_target['index']
test_target.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
max_depth_list = []
for i in range(18,35,1):
    tree_clf = DecisionTreeClassifier(max_depth=i, random_state=42)
    max_depth_list.append(tree_clf)
max_depth_list
```

```python
tree_clfs_list =[]
for i in range(17):
    tree_clfs= max_depth_list[i].fit(train_exam, train_target)
    tree_clfs_list.append(tree_clfs)
tree_clfs_list
```

```python
train_list_score = [] 
for i in range(17):
    train_list_score.append(tree_clfs_list[i].score(train_exam, train_target)) 
train_list_score
```

```python
from sklearn.metrics import accuracy_score
```

```python
com_target_list = []
for i in range(17):
    com_target = max_depth_list[i].predict(test_exam)
    com_target_list.append(com_target)
com_target_list
```

```python
accuracy_score_list = []
for i in range(17):
    accuracy_score_list.append(accuracy_score(test_target, com_target_list[i]))
accuracy_score_list
```
