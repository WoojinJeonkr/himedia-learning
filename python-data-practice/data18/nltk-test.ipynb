# 형태소 분석
형태소(morpheme)는 언어학에서 일정한 의미가 있는 가장 작은 말의 단위를 뜻한다.    
보통 자연어 처리에서는 토큰으로 형태소를 이용한다.    
형태소 분석(morphological analysis)이란 단어로부터 어근, 접두사, 접미사, 품사 등 다양한 언어적 속성을 파악하고 이를 이용하여 형태소를 찾아내거나 처리하는 작업이다.    
형태소 분석의 예로는 다음과 같은 작업이 있다.
  - 어간 추출(stemming)
  - 원형 복원(lemmatizing)
  - 품사 부착(Part-Of-Speech tagging)

```python
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
st1 = PorterStemmer()
st2 =  LancasterStemmer()
```

```python
words = ["fly", "flies", "flying", "flew", "flown"]
```

```python
print('Porter >> ', [st1.stem(w) for w in words])
print('Lang >> ', [st2.stem(w) for w in words])
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lm = WordNetLemmatizer()
```

```python
print('wordNet >> ',  [lm.lemmatize(w, pos="v") for w in words])
```

- DT: 관형사

# 불용어 처리와 단어 추출

```python
from nltk import FreqDist
```

```python
emma_tokens = pos_tag(retokenize.tokenize(emma_raw))
```

```python
emma_tokens
```

```python
# 불용어 작성
stopwords=["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear", "to", "the", "and", "of", "I", 'was', "nor"]
```

```python
emma_tokens
```

```python
# for문이 돌음 -> w[0]가 stopwords에 없으면서 'NNP'인 경우 --> w[0]를 w에 담음
emma_token2 = [w[0] for w in emma_tokens if w[1] == 'NNP' and w[0] not in stopwords]
```

```python
emma_token2[:3]
```

```python
fd_names = FreqDist(emma_token2)
```

```python
fd_names.most_common(5) # most_common : 자주 등장하는 단어의 순위를 추출
```
