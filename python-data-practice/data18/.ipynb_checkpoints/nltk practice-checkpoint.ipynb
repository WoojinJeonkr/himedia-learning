# 뉴스 기사 가져오기

### * 뉴스 출처 : https://www.independent.co.uk/arts-entertainment/music/reviews/bts-review-live-la-b1972415.html

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.independent.co.uk/arts-entertainment/music/reviews/bts-review-live-la-b1972415.html'
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    content = soup.select_one('#main')
    bts = content.get_text()
    print(bts)

else : 
    print(response.status_code)
```

## 기사 토큰화하기

```python
from nltk.tokenize import sent_tokenize
```

```python
print(sent_tokenize(bts[:1000])[3])
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(bts[50:100])
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
retokenize = RegexpTokenizer("[\w]+")
```

```python
retokenize.tokenize(bts[50:100])
```

## 품사

```python
from nltk.tag import pos_tag
```

```python
tagged_list = pos_tag(word_tokenize(bts))
```

```python
tagged_list[:3]
```

## 불용어와 단어추출

```python
from nltk import FreqDist
```

```python
bts_tokens = pos_tag(retokenize.tokenize(bts))
```

```python
bts_tokens
```

```python
# 불용어 불러오기
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
```

```python
# for문이 돌음 -> w[0]가 stopwords에 없으면서 'NNP'인 경우 --> w[0]를 w에 담음
bts_tokens2 = [w[0] for w in bts_tokens if w[1] == 'NNP' and w[0] not in stopwords]
```

```python
bts_tokens2[:5]
```

```python
fd_names = FreqDist(bts_tokens2)
```

```python
fd_names.most_common(5)
```

## wordcloud 작성하기

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```

```python
# 워드클라우드 이미지로 저장
wc.to_file(filename="bts_news_nltk.png")
```
