```python
ranker=[]
for i in range(1, 11):
    con=request.urlopen('https://www.op.gg/leaderboards/tier?page='+str(i)+'&region=kr')
    doc = BeautifulSoup(con, 'html.parser')
    ranker += doc.select('strong')
    ranker.pop()
ranker
```

```python
len(ranker)
```

```python
name_list = rank_top + ranker
name_list
```

```python
len(name_list)
```

```python
rank_list = []
for i in range(1, 11):
    con=request.urlopen('https://www.op.gg/leaderboards/tier?page='+str(i)+'&region=kr')
    doc = BeautifulSoup(con, 'html.parser')
    rank_list += doc.select('.rank')
rank_list
```

```python
len(rank_list)
```

```python
level_list = []
for i in range(1, 11):
    con=request.urlopen('https://www.op.gg/leaderboards/tier?page='+str(i)+'&region=kr')
    doc = BeautifulSoup(con, 'html.parser')
    level_list += doc.select('.level')
level_list
```

```python
len(level_list)
```

```python
win = doc.select('.w')
win
```

```python
win_list = []
for i in range(1, 11):
    con=request.urlopen('https://www.op.gg/leaderboards/tier?page='+str(i)+'&region=kr')
    doc = BeautifulSoup(con, 'html.parser')
    win_list += doc.select('.w')
win_list
```

```python
len(win_list)
```

```python
ranking = []
for one in rank_list:
    ranking.append(one.text)
    
name = []
for one in name_list:
    name.append(one.text)

level = []
for one in level_list:
    level.append(one.text)
for i in range(1, 5):
    temp=level[i].split('.')
    level[i]=temp[1]
    
win = []
for one in win_list:
    win.append(one.text)
level
```

```python
opgg_file = open('miniproject.csv', 'a', encoding='utf-8-sig')
# rank,name,level,win
```

```python
for i in range(1000):
    opgg_file.write(ranking[i] + ',' +
                    name[i] + ',' +
                    level[i] + ',' +
                    win[i]+ '\n')
```

```python
opgg_file.close()
```
