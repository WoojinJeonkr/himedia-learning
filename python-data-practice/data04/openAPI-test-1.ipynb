```python
# 한국언론진흥재단_뉴스빅데이터_메타데이터_코로나
import requests

header_data = {
    'accept': 'application/json',
    'Authorization': '공공데이터 포털에서 발급받은 일반 인증키(Decoding)',
}

param_data = (
    ('page', '1'),
    ('perPage', '2'),
    ('serviceKey', '공공데이터 포털에서 발급받은 일반 인증키(Decoding)'),
)

url = 'https://api.odcloud.kr/api/15069309/v1/uddi:973ad0df-617a-4565-8e82-8ea7869e75d4'
## decode key가 적용되었음.
## api마다 연결이 되지 않는다고 한다면, encode, decode를 바꾸어 테스트해볼 것.
response = requests.get(url, headers=header_data, params=param_data)
response
```

```python
# https://api.odcloud.kr/api/15069309/v1/uddi:973ad0df-617a-4565-8e82-8ea7869e75d4 # 주소
# ?
# page=1 # 페이지
# &
# perPage=2 # 페이지
# &
# serviceKey=공공데이터 포털에서 발급받은 일반 인증키(Encoding) # 인증키

# curl -X GET "https://api.odcloud.kr/api/15069309/v1/uddi:973ad0df-617a-4565-8e82-8ea7869e75d4 # 주소
# ?
# page=1 # 페이지
# &
# perPage=2 # 페이지
# &
# serviceKey=공공데이터 포털에서 발급받은 일반 인증키(Encoding)" # 인증키
# -H  
# "accept: application/json" # 요청 타입
# -H  
# "Authorization: 공공데이터 포털에서 발급받은 일반 인증키(Decoding)" # 인증키

# response 유형
# 200 : 성공적으로 수행됨
# 401 : 인증 정보가 정확 하지 않음
# 500 : API 서버에 문제가 발생하였음
```
