```python
import requests

header_data = {
    'accept': 'application/json',
    'Authorization': '공공데이터포털에서 발급받은 인증키(Decoding)',
}

param_data = (
    ('page', '1'),
    ('perPage', '10'),
    ('serviceKey', '공공데이터포털에서 발급받은 인증키(Decoding)'),
)

url = 'https://api.odcloud.kr/api/15067508/v1/uddi:6a073e23-3da9-4af5-8a3d-da2aa2008e6b'
## decode key가 적용되었음.
## api마다 연결이 되지 않는다고 한다면, encode, decode를 바꾸어 테스트해볼 것.
response = requests.get(url, headers=header_data, params=param_data)
response
```

```python
json_data = response.json()
json_data
```

```python
import pickle
```

```python
with open('water-pollution-gokr.json', 'wb') as fp:
    pickle.dump(json_data, fp) 
```

```python
with open('water-pollution-gokr.json', 'rb') as fp:
    json_read_data = pickle.load(fp) # fp라는 string을 통해서 파일을 읽어옴
```

```python
data_values = json_data.get('data')
data_values
```

```python
total_row=[]
for no in range(10):
    agency = data_values[no]['등록기관']
    rg_date= data_values[no]['등록일시']
    delete = data_values[no]['삭제여부']
    df_addr = data_values[no]['상세주소']
    river = data_values[no]['수계']
    md_date = data_values[no]['수정일자']
    notice = data_values[no]['신고일시']
    receipt = data_values[no]['접수일시']
    addr = data_values[no]['주소']
    apply = data_values[no]['지원여부']
    attach = data_values[no]['첨부파일']
    one_row = []
    one_row.append(agency)
    one_row.append(rg_date)
    one_row.append(delete)
    one_row.append(df_addr)
    one_row.append(river)
    one_row.append(md_date)
    one_row.append(notice)
    one_row.append(receipt)
    one_row.append(addr)
    one_row.append(apply)
    one_row.append(attach)
    print(one_row)
    total_row.append(one_row)
    print('전체 자료 수 >>> ', len(total_row))
```

```python
with open('water-sub-data.json','wb') as fp:
    pickle.dump(total_row, fp)
```

```python
with open('water-sub-data.json','rb') as fp:
    json_sub_data = pickle.load(fp)
json_sub_data
```
