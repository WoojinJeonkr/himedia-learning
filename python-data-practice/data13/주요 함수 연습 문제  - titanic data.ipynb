##### **문제 1. p-class column update => 1: low, 2-3: middle, 4-5: high로 표시하세요**

```python
def pclass_update(x):
    if x == 1:
        return 'low'
    elif 2<= x <=3:
        return 'middle'
    else:
        return 'high'
```

```python
titanic['pclass'] = titanic['pclass'].apply(pclass_update)
```

```python
titanic
```

##### **문제 2. name으로 파생변수 name_result를 만드세요. 글자수를 세어 15 이상이면 long, 나머지는 short으로 표시하세요**

```python
len_name =  lambda x: 'long' if len(x) >= 15 else 'short'
```

```python
titanic['name_result'] = list(map(len_name, titanic['name']))
titanic
```

##### **문제 3. survived가 1이면 생존, 0이면 사망으로 판단하여 survived column을 update하세요**

```python
judge_survived = lambda x: '생존' if x == 1 else '사망'
```

```python
titanic['survived'] = titanic['survived'].apply(judge_survived)
titanic
```

##### **문제 4. age 10대, 20대, 30대, 40대, 50대, 60대 이상으로 age column을 update하세요**

```python
def classify_age(x):
    if x >= 0:
        return '10대 미만'
    elif x >= 10:
        return '10대'
    elif x >= 20:
        return '20대'
    elif x >= 30:
        return '30대'
    elif x >= 40:
        return '40대'
    elif x >= 50:
        return '50대'
    else:
        return '60대'
```

```python
titanic['age'] = titanic['age'].apply(classify_age)
titanic
```

##### **문제 5. name에 Mr, Dr을 포함하고 있다면 '남자', Miss를 포함하고 있으면 '여자', 나머지는 '기타'로 파생변수 name_result2를 만들고,**
##### 　　　**&nbsp;각각을 카운트하여 남자와 여자의 수를 구해보세요**

```python
name_gender = lambda x: '남자' if 'Mr' in x else '남자' if 'Dr' in x else '여자' if 'Miss' in x else '기타'
```

```python
titanic['name_result2'] = titanic['name'].apply(name_gender)
titanic
```

```python
male = titanic['name_result2'][titanic['name_result2'] == '남자'].count()
male
```

```python
female = titanic['name_result2'][titanic['name_result2'] == '여자'].count()
female
```
