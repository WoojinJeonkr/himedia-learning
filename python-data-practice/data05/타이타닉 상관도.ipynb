```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
corr = df.corr()
corr
```

```python
cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']
```

```python
sns.set(font_scale=1.5) # font 크기 설정
sns.heatmap(corr.values,
           cbar = True, # 우측 바 표시
           annot = True, # 값 표시
           square = True, # 모양
            fmt = '.2f', # 소수점 n번째 자리까지 표시
           xticklabels = cols, # x축 라벨을 cols로
            yticklabels = cols # y축 라벨을 cols로
           ) # corr의 값을 heatmap으로 시각화
sns.set(rc = {'figure.figsize':(20,15)})
```

```python
sns.pairplot(df, height=2.5)
plt.show()
```
