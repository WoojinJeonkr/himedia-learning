```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
corr = df.corr()
corr
```

```python
cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']
```

```python
sns.set(font_scale=1.5) # font í¬ê¸° ì„¤ì •
sns.heatmap(corr.values,
           cbar = True, # ìš°ì¸¡ ë°” í‘œì‹œ
           annot = True, # ê°’ í‘œì‹œ
           square = True, # ëª¨ì–‘
            fmt = '.2f', # ì†Œìˆ˜ì  në²ˆì§¸ ìžë¦¬ê¹Œì§€ í‘œì‹œ
           xticklabels = cols, # xì¶• ë¼ë²¨ì„ colsë¡œ
            yticklabels = cols # yì¶• ë¼ë²¨ì„ colsë¡œ
           ) # corrì˜ ê°’ì„ heatmapìœ¼ë¡œ ì‹œê°í™”
sns.set(rc = {'figure.figsize':(20,15)})
```
