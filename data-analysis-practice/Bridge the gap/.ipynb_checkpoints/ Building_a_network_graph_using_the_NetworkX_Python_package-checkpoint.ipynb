# NetworkX 파이썬 패키지를 이용한 네트워크 그래프 작성

## 1. NetworkX 파이썬 패키지란?

- 그래프와 네트워크를 연구하기 위한 Python 라이브러리
- BSD-new 라이선스 하에 출시된 무료 소프트웨어
- 1000만 노드와 1000만 에지를 초과하는 그래프와 같은 대규모 실제 그래프 작업에 적합
- 네트워크 및 소셜 네트워크 분석을 위한 합리적으로 효율적이고 확장성이 뛰어나고 이식성이 높은 프레임워크

## 2. 그래프 작성하기

### 01. 데이터 분석

```python
# 데이터 출처: https://www.kaggle.com/datasets/uciml/zoo-animal-classification?resource=download&select=zoo.csv
```

```python
# 라이브러리 추가
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import networkx as nx
```

```python
zoo = pd.read_csv('./zoo.csv')
zoo
```

```python
# networkx 그래프에 그려질 class_type 변경
zoo.loc[(zoo.class_type == 1), 'class_type'] = 'Mammal'
zoo.loc[(zoo.class_type == 2), 'class_type'] = 'Bird'
zoo.loc[(zoo.class_type == 3), 'class_type'] = 'Reptile'
zoo.loc[(zoo.class_type == 4), 'class_type'] = 'Fish'
zoo.loc[(zoo.class_type == 5), 'class_type'] = 'Amphibian'
zoo.loc[(zoo.class_type == 6), 'class_type'] = 'Bug'
zoo.loc[(zoo.class_type == 7), 'class_type'] = 'Invertebrate'
```

```python
zoo
```

## 02. 네트워크 그래프 그리기

```python
g = nx.Graph()
g = nx.from_pandas_edgelist(zoo, source = 'animal_name', target = 'class_type')
print(nx.info(g))

plt.figure(figsize=(18, 18))
plt.rc('NanumGothic')
pos = nx.spring_layout(g, k = 0.3)
nx.draw_networkx(g, pos, node_size = 20, node_color = 'blue', font_size = 16, font_weight='bold')
plt.show()
```
