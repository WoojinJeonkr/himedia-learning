# **정현파 조합**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1024
T = 1.0 / 44100.0
f1 = 697
f2 = 1209
```

```python
t = np.linspace(0.0, N*T, N)
```

```python
y1 = 1.1 * np.sin(2 * np.pi * f1 * t)
y2 = 0.9 * np.sin(2 * np.pi * f2 * t)
```

```python
y = y1 + y2
```

```python
plt.subplot(311)
plt.plot(t, y1)
plt.title(r"$1.1\cdot\sin(2\pi\cdot 697t)$")
plt.subplot(312)
plt.plot(t, y2)
plt.title(r"$0.9\cdot\sin(2\pi\cdot 1209t)$")
plt.subplot(313)
plt.plot(t, y)
plt.title(r"$1.1\cdot\sin(2\pi\cdot 697t) + 0.9\cdot\sin(2\pi\cdot 1209t)$")
plt.tight_layout()
plt.show()
```

# **고속 퓨리에 변환(Fast Fourier Transform, FFT)**

```python
y2 = np.hstack([y, y, y])
```

```python
plt.subplot(211)
plt.plot(y2)
plt.axvspan(N, N * 2, alpha=0.3, color='green')
plt.xlim(0, 3 * N)
plt.show()
```

```python
plt.subplot(212)
plt.plot(y2)
plt.axvspan(N, N * 2, alpha=0.3, color='green')
plt.xlim(900, 1270)
plt.show()
```

# **피리오도그램(periodogram) 그리기**

```python
from scipy.fftpack import fft
```

```python
yf = fft(y, N)
```

```python
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
```

```python
plt.stem(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlim(0, 3000)

plt.show()
```

# **DCT(Discrete Cosine Transform)**

```python
from scipy.fftpack import dct
```

```python
dct_type = 2
```

```python
yf2 = dct(y, dct_type, N)
```

```python
plt.subplot(311)
plt.stem(np.real(yf))
plt.title("DFT real part")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(312)
plt.stem(np.imag(yf))
plt.title("DFT imaginary part")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(313)
plt.stem(np.abs(yf2))
plt.title("DCT")

plt.tight_layout()
plt.show()
```

# **스펙트럼(power spectrum)**

```python
import scipy.signal
```

```python
f, P = scipy.signal.periodogram(y, 44100, nfft=2**12)
```

```python
plt.subplot(211)
plt.plot(f, P)
plt.xlim(100, 1900)
plt.title("linear scale")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(212)
plt.semilogy(f, P)
plt.xlim(100, 1900)
plt.ylim(1e-5, 1e-1)
plt.title("log scale")

plt.tight_layout()
plt.show()
```

# **STFT(Short-Time Fourier Transform)**

```python
import librosa
import librosa.display
```

```python
D = np.abs(librosa.stft(y))
```

```python
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), y_axis='linear', x_axis='time')
plt.title('Dual Tone')
plt.ylim(0, 4000)
plt.show()
```

```python
import scipy.io.wavfile
```

```python
sr_octave, y_octave = scipy.io.wavfile.read("octave.wav")
```

```python
D_octave = np.abs(librosa.stft(y_octave))
```

```python
librosa.display.specshow(librosa.amplitude_to_db(D_octave, ref=np.max), sr=sr_octave, y_axis='linear', x_axis='time')
plt.title('Octave')
plt.ylim(0, 2000)
plt.show()
```
