    - 대립가설(연구가설) : '차이가 있다.' 또는 '효과가 있다.'
    - 연구 가설 제시 시, 귀무가설을 통해서 가설 제시
    - 귀무가설이 낮을 경우, 대립가설이 높을 경우 채택 

## 3. 유의수준 결정

- 유의수준(α) : 가설 채택 또는 기각 기준, 임계값, P-value를 얼마나 유의하다고 결정할 임계값
- 알파(α) : 가설을 신뢰할 수 없는 확률(5%) <-> 신뢰수준(1- α) : 가설을 신뢰할 수 있는 확률(95%)
- 분석 결과 유의수준 이내 🡺 가설 채택(그렇지 않으면 기각)
- p값 (p-value) : 검정통계량에 의해서 구해진 값 
- α > P값 : 귀무가설 기각(대립가설 채택)   
  α ≦ P값 : 귀무가설 채택
