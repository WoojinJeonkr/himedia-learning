## 2. 확률 분포 (Probability distribution)

- 확률변수 : 일정한 확률(값)을 갖는 변수
- 확률변수 X가 특정한 값을 가질 확률을 나타낸 분포
- 이산확률분포(이산확률변수) : 확률변수 X가 가질 수 있는 값이 셀 수 있는 집합  
ex) 주사위를 던졌을 때 나오는 수, 동전의 앞면/뒷면(이항분포)
- 연속확률분포(연속확률변수) : 확률변수 X가 어떤 구간 안에 있는 모든 실수값을 갖는 분포   
ex) 전체 남학생의 키, 몸무게

## 7. 검정통계량 (Test statistic)

- 가설 검정 위해 수집된 표본 자료를 계산한 통계량
- 가설검정에서 임계값과 비교하여 가설검정에 사용되는 통계량
- 유의수준 α의 값와 비교하여 귀무가설 기각/채택
- T검정 t값, 분산분석/회귀분석 F값, 카이제곱 X2값, 상관분석 r값

### 7-1. 문제   
연구가설(H1) : ‘학력수준에 따라 제품만족도에 차이가 있다.’를 검정하기 위해서 독립표본 T검정을 수행하였다. 이때 유의수준은 α=0.05로 결정하였다.   
검정 결과 검정통계량 t값이 10.652, 유의확률 p값이 0.012가 나왔다고 가정한다면 귀무가설은 기각되는가? 채택되는가?

### 정답
검정통계량 t=10.652값은 유의확률 p=0.012이다.   
유의수준 α=0.05 수준에서 귀무가설 기각(p<α)   
학력수준에 따라 제품만족도에 유의미한 차이가 있다고 볼 수 있다

- 신뢰수준에 따라서 신뢰구간이 결정된다.

- Z값 = 1.96에 해당하는 면적이 0. 475이고 좌우대칭인 확률은 0.95(95%)에 해당한다.

- T와 F분포도 좌우대칭인 정규분포형태이기 때문이다.

## 10. 기각역 vs 채택역

- 임계값(Critical value) : 귀무가설 채택 or 기각 기준점
- 채택역(Acceptance region) : 임계값 기준 채택(귀무가설) 범위 = 신뢰구간(가설을 채택할 수 있는 영역)
- 기각역(Critical region) : 기각 범위

```python
Image('./임계값, 기각역, 채택역, 양측검정.PNG')
```

### 10-1. 양측검정 (2-sided test)
알파값이 0.05이면 평균을 중심으로 0.025씩 양쪽으로 분류되어 채택역이 형성된다. 이러한 검정을 양측검정이라고 한다.    
H0 : 성별에 따라 만족도에 차이가 없다.(같다)   
H1 : 성별에 따라 만족도에 차이가 있다.(같지 않다)  

```python
Image('./양측검정.PNG')
```

### 10-2. 단측검정 (1-sided test)
방향(우열) 있는 단측가설 검정   
- H0 : 1일 생산되는 불량품의 개수는 평균 30개 이다.(μ=30)   
- H1 : 1일 생산되는 불량품의 개수는 평균 30개 이하이다.(μ<30) ▶ 왼쪽 단측검정
