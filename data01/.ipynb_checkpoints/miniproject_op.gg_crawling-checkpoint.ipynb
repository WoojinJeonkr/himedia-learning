```python
from urllib import request
```

```python
from bs4 import BeautifulSoup
```

```python
!dir
```

```python
con = request.urlopen('https://www.op.gg/leaderboards/tier?region=kr')
con
```

```python
con.status
```

```python
doc = BeautifulSoup(con, 'html.parser')
doc
```

```python
rank = doc.select('.rank')
rank
```

```python
rank_list = []
for one in rank:
    rank_list.append(one.text)
rank_list
```

```python
rank_top = doc.select('.name')
rank_top
```

```python
len(rank_top)
```

```python
rank_top_list = []
for one in rank_top:
    rank_top_list.append(one.text)
rank_top_list
```

```python
ranker = doc.select('strong')
ranker.pop()
ranker
```

```python
len(ranker)
```

```python
ranker_list = []
for one in ranker:
    ranker_list.append(one.text)
ranker_list
```

```python
name_list = rank_top_list + ranker_list
name_list
```

```python
level = doc.select('.level')
level
```

```python
level_list = []
for one in level:
    level_list.append(one.text)
level_list
```

```python
win = doc.select('.w')
win
```

```python
win_list = []
for one in win:
    win_list.append(one.text)
win_list
```

```python
opgg_file = open('miniproject.csv', 'a', encoding='utf-8-sig')
# rank,name,level,win
```

```python
for i in range(100):
    opgg_file.write(rank_list[i] + ',' +
                    name_list[i] + ',' +
                    str(level_list[i]) + ',' +
                    str(win_list[i])+ '\n')
```

```python
opgg_file.close()
```
