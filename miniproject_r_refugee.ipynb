```python
col = ['망명년도', '망명지역', '실거주지', '난민 지위', '총 보류 누적 사람 수(시작 연도 기준으로 시작)', 
           '유엔난민기구 지원 사람 수(시작 연도 기준으로 시작)', '전쟁 난민 신청 수', '전쟁 난민 결정 수',
           '전쟁 난민 보류 수', '전쟁 난민 거부 수', '전쟁 난민 철회 수', '전쟁 난민 인원 현황',
          '총 보류 누적 사람 수(시작 연도 기준으로 종료)', '유엔난민기구 지원 사람 수(시작 연도 기준으로 종료)']
```

```python
refugee = pd.read_csv("./asylum_seekers.csv",header=1, low_memory=False, names=col)
```

```python
refugee = refugee.replace('*',0)
```

```python
refugee2 = refugee.fillna(0)
```

```python
refugee2
```

```python
refugee2 = refugee2.astype({'전쟁 난민 인원 현황':'int'})
```

```python
refugee_region = refugee2[['망명지역', '전쟁 난민 인원 현황']]
refugee_region
```

```python
sort_refugee = refugee_region.pivot_table(index='망명지역', aggfunc=np.sum)
sort_refugee.reset_index(inplace=True)
sort_refugee
```

```python
resion_refugee = []
r_refugee = ['Germany', 'Algeria', 'Angola', 'Iran (Islamic Rep. of)', 'Yemen', 'Sweden',
            'Hungary', 'United Kingdom', 'Malaysia', 'Rep. of Korea'] 
for i in range(len(r_refugee)):
    resion_refugee.append(sort_refugee[sort_refugee['망명지역']==r_refugee[i]]['전쟁 난민 인원 현황'].sum())
resion_refugee
```

```python
total_r_refugee = pd.DataFrame(data = resion_refugee,
                               index=r_refugee,
                               columns=['전쟁 난민 총 인원'])
total_r_refugee.index.name = '망명지역'
total_r_refugee
```

```python
total_r_refugee.to_csv('total_r_refugee.csv', encoding='UTF-8-sig')
```
