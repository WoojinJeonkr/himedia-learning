```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
import joblib
```

```python
pos_dict = joblib.load('pos_dict.pkl')
words = [n for n in pos_dict if len(n) > 1]
pos_count = Counter(words)
wc = WordCloud(font_path='malgun', width=600, height=600, scale=2.0, max_font_size=250, background_color='white')
gen = wc.generate_from_frequencies(pos_count)
plt.figure()
plt.imshow(gen)
```

```python
neg_dict = joblib.load('neg_dict.pkl')
words = [n for n in neg_dict if len(n) > 1]
neg_count = Counter(words)
wc = WordCloud(font_path='malgun', width=600, height=600, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(neg_count)
plt.figure()
plt.imshow(gen)
```
